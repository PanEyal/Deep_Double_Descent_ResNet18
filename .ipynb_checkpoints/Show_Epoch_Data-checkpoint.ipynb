```python
import numpy as np

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.utils.data
from torch.utils.data import Dataset

import torchvision
from torchvision import datasets, transforms

import matplotlib.pyplot as plt

from tqdm import tqdm

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
train_error_0 = torch.load('./Epochs_Output/0_noise_train_error_list.pt')
train_loss_0 = torch.load('./Epochs_Output/0_noise_train_loss_list.pt')
test_error_0 = torch.load('./Epochs_Output/0_noise_test_error_list.pt')

train_error_10 = torch.load('./Epochs_Output/10_noise_train_error_list.pt')
train_loss_10 = torch.load('./Epochs_Output/10_noise_train_loss_list.pt')
test_error_10 = torch.load('./Epochs_Output/10_noise_test_error_list.pt')

train_error_20 = torch.load('./Epochs_Output/20_noise_train_error_list.pt')
train_loss_20 = torch.load('./Epochs_Output/20_noise_train_loss_list.pt')
test_error_20 = torch.load('./Epochs_Output/20_noise_test_error_list.pt')
```

```python
plt.plot(range(len(train_error_0)), train_error_0, label="0% noise")
plt.plot(range(len(train_error_10)), train_error_10, label="10% noise")
plt.plot(range(len(train_error_20)), train_error_20, label="20% noise")
plt.xlabel('Epochs')
plt.ylabel('Train Error')
plt.xscale('log')
plt.legend()
```

```python
plt.plot(range(len(train_loss_0)), train_loss_0, label="0% noise")
plt.plot(range(len(train_loss_10)), train_loss_10, label="10% noise")
plt.plot(range(len(train_loss_20)), train_loss_20, label="20% noise")
plt.xlabel('Epochs')
plt.ylabel('Train Loss')
plt.xscale('log')
plt.legend()
```

```python
plt.plot(range(len(test_error_0)), test_error_0, label="0")
plt.plot(range(len(test_error_10)), test_error_10, label="10")
plt.plot(range(len(test_error_20)), test_error_20, label="20")
plt.xlabel('Epochs')
plt.ylabel('Test Error')
plt.xscale('log')
plt.legend()
```
